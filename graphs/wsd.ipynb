```python
import matplotlib.pyplot as plt

import wandb
```

```python
api = wandb.Api()
```

```python
run_wsd = api.run("pufanyi/sc4001/0js1pr83")
run_cos = api.run("pufanyi/sc4001/jhn7i0f0")
```

```python
train_loss_wsd = run_wsd.history(keys=["train/loss"])
train_loss_cos = run_cos.history(keys=["train/loss"])
eval_loss_wsd = run_wsd.history(keys=["eval/loss"])
eval_loss_cos = run_cos.history(keys=["eval/loss"])
train_learning_rate_wsd = run_wsd.history(keys=["train/lr"])
train_learning_rate_cos = run_cos.history(keys=["train/lr"])
```

```python
# Morandi color palette (muted, elegant colors)
colors = {
    "with_trans": "#8B9FA8",  # Dusty blue
    "without_trans": "#C7A5A5",  # Dusty rose
}

# Set up the plot with paper-quality settings
plt.figure(figsize=(10, 6), dpi=300)
plt.style.use("seaborn-v0_8-darkgrid")

# Plot training losses (solid lines)
plt.plot(
    train_loss_wsd["_step"],
    train_loss_wsd["train/loss"],
    color=colors["with_trans"],
    linewidth=2.5,
    linestyle="-",
    label="Train Loss (WSD)",
    alpha=0.9,
)

plt.plot(
    train_loss_cos["_step"],
    train_loss_cos["train/loss"],
    color=colors["without_trans"],
    linewidth=2.5,
    linestyle="-",
    label="Train Loss (Cosine)",
    alpha=0.9,
)

# Plot evaluation losses (dashed lines, slightly darker)
plt.plot(
    eval_loss_wsd["_step"],
    eval_loss_wsd["eval/loss"],
    color=colors["with_trans"],
    linewidth=2.5,
    linestyle="--",
    label="Eval Loss (WSD)",
    alpha=0.9,
)

plt.plot(
    eval_loss_cos["_step"],
    eval_loss_cos["eval/loss"],
    color=colors["without_trans"],
    linewidth=2.5,
    linestyle="--",
    label="Eval Loss (Cosine)",
    alpha=0.9,
)

# Apply log scale to y-axis for better visualization
plt.yscale("log")

# Customize the plot for paper quality
plt.xlabel("Training Step", fontsize=14, fontweight="bold")
plt.ylabel("Loss (log scale)", fontsize=14, fontweight="bold")
plt.title(
    "Effect of Learning Rate Scheduler on Training Dynamics",
    fontsize=16,
    fontweight="bold",
    pad=20,
)

# Legend with better positioning
plt.legend(
    loc="upper right", fontsize=11, framealpha=0.95, edgecolor="gray", fancybox=True
)

# Grid for better readability
plt.grid(True, alpha=0.3, linestyle=":", linewidth=0.8, which="both")

# Tight layout for better spacing
plt.tight_layout()

# Save the figure in high quality
plt.savefig(
    "loss_curves_lr.png",
    dpi=300,
    bbox_inches="tight",
    facecolor="white",
    edgecolor="none",
)
plt.savefig(
    "loss_curves_lr.pdf", bbox_inches="tight", facecolor="white", edgecolor="none"
)

plt.show()
```

```python
# Morandi color palette (muted, elegant colors)
colors = {
    "wsd": "#8B9FA8",  # Dusty blue
    "cosine": "#C7A5A5",  # Dusty rose
}

# Set up the plot with paper-quality settings
plt.figure(figsize=(10, 6), dpi=300)
plt.style.use("seaborn-v0_8-darkgrid")

# Plot learning rate curves
plt.plot(
    train_learning_rate_wsd["_step"],
    train_learning_rate_wsd["train/lr"],
    color=colors["wsd"],
    linewidth=2.5,
    linestyle="-",
    label="Learning Rate (WSD Scheduler)",
    alpha=0.9,
)

plt.plot(
    train_learning_rate_cos["_step"],
    train_learning_rate_cos["train/lr"],
    color=colors["cosine"],
    linewidth=2.5,
    linestyle="-",
    label="Learning Rate (Cosine Scheduler)",
    alpha=0.9,
)

# Customize the plot for paper quality
plt.xlabel("Training Step", fontsize=14, fontweight="bold")
plt.ylabel("Learning Rate", fontsize=14, fontweight="bold")
plt.title(
    "Comparison of Learning Rate Schedulers: WSD vs Cosine",
    fontsize=16,
    fontweight="bold",
    pad=20,
)

# Adjust y-axis limit to prevent legend from covering the curves
plt.ylim(0, 0.000115)

# Legend with better positioning
plt.legend(
    loc="upper right", fontsize=11, framealpha=0.95, edgecolor="gray", fancybox=True
)

# Grid for better readability
plt.grid(True, alpha=0.3, linestyle=":", linewidth=0.8, which="both")

# Tight layout for better spacing
plt.tight_layout()

# Save the figure in high quality
plt.savefig(
    "learning_rate_comparison.png",
    dpi=300,
    bbox_inches="tight",
    facecolor="white",
    edgecolor="none",
)
plt.savefig(
    "learning_rate_comparison.pdf",
    bbox_inches="tight",
    facecolor="white",
    edgecolor="none",
)

plt.show()

print(
    "Figure saved as 'learning_rate_comparison.png' and 'learning_rate_comparison.pdf'"
)
```
